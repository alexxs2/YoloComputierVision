```python
import os
import cv2
from ultralytics import YOLO

# 2) Defina caminhos absolutos no Colab
video_path = '/content/gdrive/MyDrive/Yolocomputervision/data/GATO IMITANDO XJ6 [ORIGINAL](720P_HD).mp4'
video_path_out = '/content/gdrive/MyDrive/Yolocomputervision/data/GATO IMITANDO XJ6 [ORIGINAL](720P_HD)_detecção_last.mp4'
model_path = '/content/gdrive/MyDrive/Yolocomputervision/data/runs/detect/train/weights/last.pt'

# 3) Abra o vídeo para leitura e prepare o writer
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
if not ret:
    raise RuntimeError(f"Não foi possível abrir o vídeo em {video_path}")
H, W = frame.shape[:2]
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(video_path_out, fourcc, fps, (W, H))

# 4) Carregue o modelo treinado
model = YOLO(model_path)

threshold = 0.5

# 5) Loop de inferência e gravação
while ret:
    results = model(frame)[0]
    for box in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = box
        if score > threshold:
            label = results.names[int(class_id)].upper()
            # Desenha retângulo
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            # Coloca texto
            cv2.putText(
                frame,
                f"{label} {score:.2f}",
                (int(x1), int(y1) - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0, 255, 0),
                2,
                cv2.LINE_AA
            )
    out.write(frame)
    ret, frame = cap.read()

# 6) Libere recursos
cap.release()
out.release()

print(f"Vídeo processado e salvo em: {video_path_out}")
```
